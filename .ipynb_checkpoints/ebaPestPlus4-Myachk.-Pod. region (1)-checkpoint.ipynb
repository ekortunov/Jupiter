```python
# --- вычисление ABIC
def ABICcalc(x):
    cm = x[0]
    cd = x[1]
    Cm = np.diag(np.ones((m.size)))*cm
    Cd = np.diag(np.ones((g.size)))*cd
    Cdp = np.dot(G, np.dot(Cm,G.T))+Cd
    Gs = Gs_#/np.linalg.norm(Gs_)#d_ - np.dot(G,s)
    #print(np.log(np.linalg.det(Cdp)))
    ABIC = np.log(np.linalg.det(Cdp))+np.dot(Gs.T, np.dot(np.linalg.inv(Cdp), Gs))+\
    ((cm<=0)+(cd<=0)+(cm>1000))*10**8
    return ABIC
def vec2matrix(m, L=[121, 111]):
    M = np.zeros(L)
    for i in range(L[0]):
        M[i] = m[i*L[1]:(i+1)*L[1]]
    return M
#M = vec2matrix(m, L); plt.imshow(M); plt.show()
```

```python
%%time
# --- РАСЧЕТ МАТРИЦЫ КОВАРИАЦИИ ПАРАМЕТРОВ
# экспоненциальная модель вариограммы
def spherModel(h, c, c0, r):
    gamma = c0 + c*(1.5*h/r-0.5*(h/r)**3)
    gamma[h>r] = c0+c
    gamma[h<=0] = 0
    return gamma
def expModel(h, c, c0, r):
    gamma = c0+c*(1-np.exp(-h/r))
    gamma[h<=0] = 0
    return gamma
def covarian(h, c, c0, r, sgm2, mod = 'exp'):
    if mod == 'exp':
        return sgm2 - expModel(h, c, c0, r)#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
    else:
        return sgm2 - spherModel(h, c, c0, r)

def CovMatrix(L, c, c0, r, sgm2, mod = 'exp'):
    vec = []
    for i in range(L[0]):
        for j in range(L[1]):
            vec.append(np.array([i,j]))
    
    sz = L[0]*L[1]
    covMat = np.diag(np.ones(sz))
    for i in range(sz):
        #if int(i/500)*500 == i: print(i)
        for j in range(i+1, sz):
            d = vec[i] - vec[j]
            covMat[i,j] = covarian(np.array([np.dot(d,d)**0.5]), c, c0, r, sgm2, mod = mod)
            covMat[j,i] = covMat[i,j]
    return covMat

covMat = CovMatrix([121, 111], 5, 0, 15, 5) # Какую сигму взять????           
#plt.imshow(covMat-np.diag(np.ones(2500))); plt.show()   
covMat.shape
```

```python
os.chdir('C:\work\IBRAE\Калибровка\GeraModel\zven2')
def CalcHeadsRead(keys = [0.002, 0.004, 0.007, 0.009]):
    with open('monit.txt', 'r') as f:
        lines = f.readlines()
    heads = {}
    head = {}
    k=-1
    for line in lines:
        if ') name="' in line:
            if len(head)>0:
                k+=1
                heads[k]=head
                head={}
        lin = line.split(' ')
        if len(lin)==6:
            try:
                key = float(lin[0])
                val = float(lin[2])
                head[key] = val
            except: continue
    if len(head)>0:
        k+=1
        heads[k]=head
    out = []
    for k0 in heads.keys():
        for k in heads[k0].keys():
            if k in keys:
                out.append(heads[k0][k])
    return out
#CalcHeadsRead()
```

```python
# --- Считываение/запись поля проводимости ---
def readCond(name):
    with open(name, 'r') as f:
        lines = f.readlines()
    line = lines[1].replace('\n', '').replace('  ', ' ').split(' ')
    d1 = int(line[0])
    d2 = int(line[1])
    cond = np.zeros((d1,d2))
    for i in range(d2):
        lin = lines[i+2].replace('\n', '').replace('  ', ' ').split(' ')
        for j in range(d1):
            cond[j,i] = float(lin[j])
    return cond, lines[0]
cond, line0 = readCond('out1 — копия.dat')

def writeCond(name, cond, line0):
    lines = []
    lines.append(line0)
    lines.append(str(cond.shape[0])+' '+str(cond.shape[1])+'\n')
    for i in range(cond.shape[1]):
        lin = ''
        for j in range(cond.shape[0]):
            lin+=str(cond[j,i])
            if j<cond.shape[0]-1: lin+=' '
            else: lin+='\n'
        lines.append(lin)
    with open( name, 'w') as f:
        f.writelines(lines)
#writeCond('out1.dat', cond, line0)
```

```python
import copy
def singleCalc(field, name, line0):    
    writeCond(name, field, line0)    
    os.system('GeRa.bat')
    headsSim = CalcHeadsRead()
    return np.array(headsSim)
     
def deriv(gr0, field, key, dx, name):
    field1 = copy.copy(field)
    field1[key[0],key[1]] += dx
    gr = singleCalc(field1, name, line0)
    return (gr - gr0)/dx
```

```python
%%time
from sklearn.tree import DecisionTreeRegressor as tree
from sklearn.ensemble import GradientBoostingRegressor as gbr
from sklearn.ensemble import RandomForestRegressor as rf
from sklearn.neighbors import NearestNeighbors as neib
from sklearn.neighbors import KNeighborsRegressor as knn

def pilotPoints(coord, w, pilot = None):
    if pilot== None: pilot = []
    for crd in coord:
        if crd[0]>1 and [crd[0]-1, crd[1]] not in pilot: pilot.append([crd[0]-1, crd[1]])
        if crd[1]>1 and [crd[0], crd[1]-1] not in pilot: pilot.append([crd[0], crd[1]-1])
        if crd[0]<w and [crd[0]+1, crd[1]] not in pilot: pilot.append([crd[0]+1, crd[1]])
        if crd[1]<w and [crd[0], crd[1]+1] not in pilot: pilot.append([crd[0], crd[1]+1])
        
        
    return pilot

def derivApprox(pilot, Gapp, y, w, shft = 0, regressor = 'knn', max_depth = 3, n_estimators = 50, n_neighbors=2):    
    K = 2
    X = np.zeros((Gapp[0].size,2))
    for i in range(w[0]):
        X[i*w[1]:(i+1)*w[1],0] = i
        X[i*w[1]:(i+1)*w[1],1] = np.array(range(w[1]))    
    Gapp1 = np.zeros(Gapp.shape)    
    for  i in range(Gapp.shape[0]):
        if regressor == 'tree': model = tree(max_depth = 10, min_samples_leaf = 1)
        if regressor == 'boost': model = gbr(max_depth = 3, n_estimators = 50)
        if regressor == 'forest': model = rf(max_depth = 7, n_estimators = 150)
        if regressor not in ['tree', 'boost', 'forest']: 
            model = knn(n_neighbors=2, weights='distance')
        model.fit(np.array(pilot), y[:,i])
        Y = model.predict(X)
        YY = np.zeros(w)
        for j in range(shft, w[0]-1-shft):
            YY[j,shft:w[1]-shft] = Y[j*w[1]+shft:(j+1)*w[1]-shft]
        Gapp1[i,Gapp[i]==0] = YY[Gapp[i]==0]
        Gapp1[i,Gapp[i]!=0] = Gapp[i, Gapp[i]!=0]        
    return Gapp1

def derivInPP(pilot, gr, field, w, name):
    N = gr.size
    Gapp = np.zeros((N,w[0],w[1]))
    y = np.zeros((len(pilot), N))
    k=-1
    for pil in pilot:
        k+=1
        y[k,:] = deriv(gr, field, pil, 1, name)
        Gapp[:,pil[0], pil[1]] = y[k,:]
    return Gapp, y

def Jacobian(g, field, pilot, w, name, line0, shft):
    #keys = list(field.keys())
    
    #Gapp, y = derivInPP(pilot, gr, field, keys, N, w)
    Gapp, y = derivInPP(pilot, g, cond, w, name)
    return derivApprox(pilot, Gapp, y, w, shft) #derivApprox(pilot, Gapp, y, w, shft)

def meshPoints(step, shft, w, pilot):    
    for i in range(shft[0], w[0]-shft[1], step):
        for j in range(shft[2], w[1]-shft[3], step):
            if [i,j] not in pilot:
                pilot.append([i,j])
    return pilot
```
