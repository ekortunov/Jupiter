```python
import numpy as np
from scipy.optimize import fsolve
```

```python
r32 = 3.21
r34 = 2.89
t_32 = 693
t_34 = 216
Q = 0.000238
```

```python
def myFunction(z):
    nm = z[0]
    q = z[1]
  

    F = np.empty((2))
    F[0] = (nm/q)*(r34-(Q/(2*3.14*q))*np.log(1+(2*3.14*q/Q)*r34))-t_34
    F[1] = -(nm/q)*(r32+(Q/(2*3.14*q))*np.log(1-(2*3.14*q/Q)*r32))-t_32
    
    return F

zGuess = np.array([0.009,0.00001])
z = fsolve(myFunction,zGuess)
print(z)
```
